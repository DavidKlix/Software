# Python Modules

A module in python is used to hold related functions and classes.  We will only talk about how to use a module. 

If you want to write your own module, you can check out the references later.  But, remember that python is an interpreted language. This means it is relatively slow. Many of the modules that are available are written in a compiled language like C and are therefore fast.  Python can be very effective and fast when it is used as the glue to hold together calls into well written modules.

## Basic Use
To use a module you just need to import it.  Any functions or classes in the module can be used via the name as a dot prefix. To find out what is visible from the module, use the dir function.

````python
# Lets import the sympy module
import sympy

# and then use the dir function to see what it has
# Note: This really just serves to remind you what is in the
# module.  If you want to know what they are and how to use them,
# check out the documentation at https://www.scipy.org
dir(sympy)
````

````python
import sympy
# And now we can use things from the module like
# sympy.cos() and sympy.pi

# Set it up so that printing will use the best option
sympy.init_printing(use_unicode=False, wrap_line=False)


# In traditional languages, cos() is a function that would be 
# evaluated numerically.  The cos that we have imported will be
# evaluated symbolically using rules of equality
print(sympy.cos(3*sympy.pi/4))

# No rules apply for just a plain old 10
print(cos(10))

# The best printing option on display uses LaTex for formating
sympy.cos(3*sympy.pi/4)
````

## Importing for direct use
If you are someone who likes to make efficient use of their time, typing in the module name each time you want to use something from it is very annoying. Fortunately, we have a work around.  You can direct Python to import the pieces you want to use.

```Python
from module_name import name1, name2, ...
```

If you are really lazy, you can import everything, but then you have to be aware of the possibility that there is a naming clash with a function or class that you wrote.
```Python
from module_name import *
```

````python
from sympy import init_printing, Symbol, integrate

# Set it up so that printing will use the best option
init_printing(use_unicode=False, wrap_line=False)

# Define x as a symbol (its not a variable for python)
x = Symbol('x')

# Compute the integral of the polynomial function symbolically
# with respect to x (must be a symbol we know!)
integrate(x**2 + x + 1, x)
````

````python
# Another example where we import everything
from sympy import *

x = Symbol('x')

# Do symbolic integration
# Use pretty print which will use unicode
pprint(integrate(cos(x) + pi*sin(x), x))

# and again with LaTes
integrate(cos(x) + pi*sin(x), x)
````

## Creating your own module
While there is more to it, a simple way that you can create a module is to create a python file with your function definitions and then import it.  Suppose that you created a file `mywork.py` with the following contents.
```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Dec 18 13:53:48 2018

@author: ios
"""

from random import randrange
def random_choice(first, second) :
    choice = randrange(0,2)
    if choice == 0 : return first
    else : return second
    
def heads_or_tails(count):
    result = []
    for i in range(count):
        result.append(random_choice("H","T"))
    return result

# Only execute if we are the top level
if __name__ == "__main__":
    print(heads_or_tails(10))
```
You can import the file and then use the functions defined within. _Note:_  This module has code that will only be executed if the file is being run as script.  It will not be executed on import.  

Lets use mywork.

````python
import mywork

# and now we can use the its functions
print(mywork.random_choice(100, 500))

# or get the functions so you can use with out the prefix
from mywork import heads_or_tails
print( heads_or_tails(5))
````

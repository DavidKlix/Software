## Basic Plotting

In the following examples we will import functionality using shortnames because we are lazy. The `pyplot.plt` function is a simple y vs. x plotting mechanism that allows you to choose either lines, markers, or both (similar to the X-Y Scatter plot in Microsoft Excel).

Plotting a single data series is simple:

```python
plot([x], y, [fmt], data=None, **kwargs)
```

See https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot for a LOT more information

Note that the x data series is _optional_; if not provided it will default to the index of the element in the y array.

````python
import matplotlib.pyplot as plt
from math import log
# list comprehension gives the squares of each 
# Will plot with a connecting line
plt.plot([x*x for x in range(100)])
plt.ylabel('x^2')
plt.xlabel('x')
plt.show()
print('With two parameters')
# provide both an x and a y series
# also use plusses instead of a line
plt.plot([2*x for x in range(50)], [3*log(x+1) for x in range(50)], '+')
plt.show()
````

You can also plot multiple datasets on the same plot; this is very helpful when you want to compare two sets of data

````python
import matplotlib.pyplot as plt
from math import sqrt, log
vals = [x for x in range(100)]
squares = [x*x for x in range(100)]
cubes = [x*x*x for x in range(100)]
sqrts = [sqrt(x) for x in range(100)]
logs = [log(x) for x in range(1, 100)]

plt.plot(vals)
#plt.plot(squares)
#plt.plot(cubes) # plot gets hard to read with the squares and cubes in there
plt.plot(sqrts)
plt.plot(range(1, 100), logs)
plt.show()
````

Adding a legend to your plot is simple and is often required to make your results readable.

````python
import matplotlib.pyplot as plt
from math import sqrt, log
vals = [x for x in range(100)]
squares = [x*x for x in range(100)]
cubes = [x*x*x for x in range(100)]
sqrts = [sqrt(x) for x in range(100)]
logs = [log(x) for x in range(1, 100)]

plt.plot(vals, label='x')
#plt.plot(squares)
#plt.plot(cubes) # plot gets hard to read with the squares and cubes in there
plt.plot(sqrts, label='sqrt(x)')
plt.plot(range(1, 100), logs, label='log(x)')
plt.legend()
plt.show()
````
